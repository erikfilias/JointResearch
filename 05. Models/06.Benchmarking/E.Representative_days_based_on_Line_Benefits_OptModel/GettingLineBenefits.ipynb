```python
import os
import pandas as pd
```

```python
#%% Setting up the path a cases
DirName  = os.getcwd()

CaseName_Base = '9n'
CSV_name = 'oT_Result_NN_Output_'
```

```python
_path_0 = os.path.join(DirName, CaseName_Base)
```

```python
folder = _path_0+'/3.Out/0.WoParallel'
filenames = os.listdir(f"{folder}")
```

```python
df_1 = pd.read_csv(_path_0+'/3.Out'+'/0.WoParallel/'+'oT_Output_Data_'+CaseName_Base+'_Network_Existing_Generation_Full'+'.csv', index_col=[0,1,2,3,5,6])
```

```python
dfNetwork = pd.read_csv(_path_0+'/2.Par'+'/'+'oT_Data_Network_'+CaseName_Base+'.csv', index_col=[0,1,2])
```

```python
for (ni,nf,cc) in dfNetwork.index:
    if cc != 'eac1' and cc != 'eac2' and cc != 'eac3':
        df = pd.read_csv(_path_0+'/3.Out'+'/0.WoParallel/'+'oT_Output_Data_'+CaseName_Base+'_Network_Line_In_'+ni+'_'+nf+'_'+cc+'.csv', index_col=[0,1,2,3,5,6])
        print(ni,nf,cc)
        frames = [df_1, df]
        df_1   = pd.concat(frames)
```

```python
df_1 = df_1.reset_index().set_index(['Dataset'])
```

```python
df_1 = df_1.loc[['SystemCosts']]
```

```python
df_1 = df_1.reset_index().set_index(['Variable'])
```

```python
df_1
```

```python
df = df_1.loc[['vTotalGCost', 'vTotalCCost', 'vTotalECost', 'vTotalRCost']]
```

```python
df = df.reset_index()
```

```python
df
```

```python
table = pd.pivot_table(df, values='Value', index=['Period', 'Scenario', 'LoadLevel'], columns= ['Execution'], aggfunc='sum')
```

```python
table
```

```python
for (ni,nf,cc) in dfNetwork.index:
    if cc != 'eac1' and cc != 'eac2' and cc != 'eac3':
        a = 'Network_Line_In_'+ni+'_'+nf+'_'+cc
        table[a] = table['Network_Existing_Generation_Full'] - table[a]
```

```python
table = table.drop(['Network_Existing_Generation_Full'], axis=1)
```

```python
table.plot()
```
