```python
import matplotlib.pyplot as plt
import numpy as np

# List of different values to use instead of "R&D"
departments = ["R&D", "OPC"]

# Create a new figure
plt.figure(figsize=(10, 6))

# Loop through each department, get the data, and plot it
for dept in departments:
    # Get the data for each department and case
    df_utils = hm.get_utils_cm(dept, 20, case)
    np_all_util = df_utils.to_numpy().flatten()
    
    # Sort in descending order
    sorted_util = np.sort(np_all_util)[::-1]
    
    # Plot the sorted data
    plt.plot(sorted_util, label=dept)

# Add labels, title, and legend
plt.xlabel('Index')
plt.ylabel('Utility')
plt.title('Sorted Utilities for Different Departments')
plt.legend(title="Department")
plt.tight_layout()

# Show the plot
plt.show()
```
